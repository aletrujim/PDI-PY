```python
import cv2 #opencv
from matplotlib import pyplot as plt
```

```python
#cargar imagen       
image = cv2.imread(str("image.png"))

#convierte la img a RGB
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB, image)
print(img)
```

```python
#pinta la imag en el label
plt.imshow(img)
plt.show()
```

```python
#leer rango pixels       
x1 = int(10)
x2 = int(20)
y1 = int(50)
y2 = int(60)        

#leer nuev0 color       
R = int(255)             
G = int(0)
B = int(0)

#cambia color pixel
image[x1:x2, y1:y2] = [B, G, R]

#guarda la img nueva
cv2.imwrite('newimage.png', image)      

#pinta la img nueva
plt.imshow(image)
plt.show()
```
